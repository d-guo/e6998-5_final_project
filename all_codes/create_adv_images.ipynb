## Load the model 

```python
run_name = 'GANs_training_google12'
model_path = f'GANs_training/{run_name}/generator_state_dict'

gen_model = Generator().double()
gen_model.load_state_dict(torch.load(model_path))
gen_model.eval()
```

```python
save_dir = './data/google_digit_data/train_cropped_32_perturbed_run12/'
org_data_directory='./data/google_digit_data/train_cropped_32'
```

```python
from PIL import Image
import time 
from torchvision.utils import save_image

# load images and labels
files = os.listdir(org_data_directory+'/images')
files = [i.split('.')[0] for i in files if '.png' in i]
p=[]

inference_time = []

for idx, val in enumerate(files):
    
    # read image
    img = np.array(Image.open(f"{org_data_directory}/images/{val}.png"))
    img = np.moveaxis(img, -1, 0)
    
    x = (torch.tensor(img).double()).unsqueeze(0)/256
    tensor_img = x.repeat(1, 1, 1, 1)
    
    ## start model inference time
    start = time.monotonic()
    perturb = gen_model(tensor_img)
    x_perturbed = tensor_img + perturb
    x_perturbed = torch.clamp(x_perturbed, 0, 1)
    
    ## end model inference time    
    end = time.monotonic()
    inference_time.append(end-start)
    
    p.append(perturb)

    tensor_img = tensor_img.squeeze(0).permute(1, 2, 0)
    perturb = perturb.squeeze(0).permute(1, 2, 0)
    x_perturbed = x_perturbed.squeeze(0).permute(1, 2, 0)
    
    if idx % 100 == 0:
        print('original')
        figs, axs = plt.subplots(nrows=1, ncols=3, figsize=(2,6))
        axs[0].imshow(tensor_img)
        axs[1].imshow(perturb.detach().numpy())
        axs[2].imshow(x_perturbed.detach().numpy())
        axs[0].set_title("Original image",fontsize=5)
        axs[1].set_title("Perturbation",fontsize=5)
        axs[2].set_title("Perturbed image",fontsize=5)

        plt.show()
    
    # save the perturbed images to the save_dir
    save_image(x_perturbed.permute(2, 0, 1), f'{save_dir}/images/perturbed_{val}.png')
    # save the original labels to the save_dir
    shutil.copy(f"{org_data_directory}/labels/{val}.txt", f"{save_dir}/labels/perturbed_{val}.txt")
```
